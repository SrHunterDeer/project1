```python

import pandas as pd
import matplotlib.pyplot as plt
url_cleaned = 'https://raw.githubusercontent.com/SrHunterDeer/project1/main/project%20managerment/data/cleaned_data.csv'
df = pd.read_csv(url_cleaned)
print("Dữ liệu 5 dòng đầu tiên:")
display(df.head())
print("\nDanh sách các cột trong file:")
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
df['NEXT_MONTH_DEFAULT'].value_counts().sort_index().plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')
plt.title('Số lượng khách hàng vỡ nợ vs không vỡ nợ', fontsize=14, fontweight='bold')
plt.xlabel('Trạng thái (0 = không, 1 = có)', fontsize=12)
plt.ylabel('Số lượng', fontsize=12)
plt.xticks(rotation=0)
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df['AGE_YEARS'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Phân bố độ tuổi khách hàng', fontsize=14, fontweight='bold')
plt.xlabel('Tuổi', fontsize=12)
plt.ylabel('Tần suất', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['CREDIT_LIMIT_VND'], bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('Phân bố hạn mức thẻ tín dụng', fontsize=14, fontweight='bold')
plt.xlabel('Hạn mức (VND)', fontsize=12)
plt.ylabel('Tần suất', fontsize=12)
plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
